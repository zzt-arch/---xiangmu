```python
# s=pd.read_csv(r'G:\conda\anaconda\Lib\site-packages\statsmodels\tsa\tests\results\y_arma_data.csv')
# s.to_csv(r'G:\python\pytest1\项目/new.csv')
#
# s=pd.read_json(r'G:\conda\anaconda\Lib\site-packages\statsmodels\tsa\tests\results\fit_ets_results.json')
# s
# import json
import json
import pandas as pd

path = r"G:\conda\anaconda\Lib\site-packages\statsmodels\tsa\tests\results\fit_ets_results.json"

with open(path, "r", encoding="utf-8") as f:
    data = json.load(f)

rows = []
for flag, models in data.items():          # "TRUE"/"FALSE"
    for model, params in models.items():   # "AAA"/"MAA"...
        if not isinstance(params, dict):
            rows.append({"flag": flag, "model": model})
            continue

        row = {"flag": flag, "model": model}

        for k, v in params.items():
            # 把 [0.123] 这种单元素 list 变成 0.123
            if isinstance(v, list) and len(v) == 1 and not isinstance(v[0], (list, dict)):
                row[k] = v[0]
            # states 太大，别直接塞进表，做个长度摘要
            elif k == "states" and isinstance(v, list):
                row["states_len"] = len(v)
            else:
                row[k] = v

        rows.append(row)

df2 = pd.DataFrame(rows)

# 只显示你关心的几列（像Excel那样清爽）
# cols = [c for c in ["flag","model","alpha","beta","gamma","phi","states_len","initstate"] if c in df2.columns]
# df2[cols]
df2.to_excel(r'G:\python\pytest1\ets_results.xlsx', index=False)

```
